```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("/home/sudecure/LDL-Cholestrol/Sample.xlsx", header=0)
NA_tests= data.loc[data["Sonuç"].isna(), "Test Adı"].unique()

if len(NA_tests) != 0:
    print("These tests are entered with empty results: ")
    for test in NA_tests:
        print(test)


data['Sonuç'] = pd.to_numeric(data['Sonuç'], errors='coerce')
data = data.dropna(axis = 0, subset = ["Sonuç"])
data.reset_index(drop=True, inplace=True)
```
